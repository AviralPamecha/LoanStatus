```python
 def impute_LoanAmount(cols):
        LoanAmount = cols[0]
        Gender = cols[1]
        
        if pd.isnull(LoanAmount):
            if Gender == 'Male':
                return 140
            else:
                return 120
        else:
            return LoanAmount
                
            
            
        
        
```

```python
train['LoanAmount'] = train[['LoanAmount','Gender']].apply(impute_LoanAmount, axis=1)
```
